```python
import pandas as pd
import numpy as np
import time
from datetime import datetime

from google.colab import drive
drive.mount('/gdrive', force_remount=True)
```

```python
# 5개의 분리된 엑셀 파일을 불러온다
input_ = []
s_time = time.time()
for i in range(5):
  input_.append(pd.read_excel('/gdrive/My Drive/input_'+str(i+1)+'.xlsx'))
print('Elapsed :', time.time() - s_time)
```

```python
# 전처리 1 : (연월종목 X 입력변수) 의 2차원 Dataframe 생성

# 5개 각각의 엑셀 파일을 2차원 dataframe으로 정리한 뒤 5개를 1개로 병합
num_date = 58

new_input = []
for input_data in input_:

  for i in range(1, num_date): # 날짜 -> string
    input_data['Code'][i] = input_data['Code'][i].strftime('%Y%m')[2:]

  # 날짜 리스트
  date = input_data['Code'].tolist()[1:]

  num_input = 28 # 입력 변수 갯수
  num_stock = (input_data.shape[1] - 1)//num_input # 한 엑셀 파일당 종목 갯수

  # 종목별 데이터 쪼개기
  input_data_list = []
  for i in range(num_stock):
    input_data_list.append(input_data.iloc[:, num_input*i+1 : num_input*(i+1)+1])
  
  # 입력 변수 리스트
  input_list = input_data_list[0].loc[0,:].tolist()

  # 종목 당 (연월종목 X 입력변수) Dataframe 구성
  for i in range(num_stock):
    # 종목 코드 추출
    stock_code = input_data_list[i].columns[0]

    # 날짜 + 종목의 새 Index 구성
    new_date = []
    
    for t in date:
      new_date.append(t + stock_code)

    # Column 설정
    input_data_list[i].columns = input_list
    input_data_list[i].drop([0], inplace=True)

    # Index 설정
    input_data_list[i].index = new_date
      
    # Dataset X, y 시간 조정
    # 주식 종목 당 X(t) <-> y(t+1) 이 연결되도록 한다.
    # -> X : 가장 최근 값 삭제 / y : 가장 과거 값 삭제
    
    output_data = input_data_list[i][['월간수익률']][1:]
    input_data_list[i] = input_data_list[i][:-1]
    input_data_list[i]['월간수익률(t+1)'] = output_data['월간수익률'].tolist()
    
  # 종목 Dataframe 결합
  new_input.append(pd.concat(input_data_list))

# 엑셀 별 Dataframe 결합
all_input = pd.concat(new_input)
all_input
```

```python
# 전처리 2: 결측 값 제거
# y 값 (월간수익률)을 다른 값으로 채워넣기엔 성능 저하가 예측되기에 제거한다.

# (월간수익률) 이 NaN 인 Row 제거
all_input = all_input[pd.notnull(all_input['월간수익률'])]
all_input = all_input[pd.notnull(all_input['월간수익률(t+1)'])]

# 종목 종류 수 재설정
num_stock = all_input.shape[0]//num_date 
all_input.isnull().sum() # Column 당 NaN 값 갯수
```

```python
# NaN 값이 너무 많은 Column 제거
all_input.drop(['비지배주주지분'], axis='columns', inplace=True)
num_input -= 1 # Input 수 재설정

# 2번째로 NaN 값이 제일 많은 Column (기타포괄이익누계액)이 NaN인 Row 제거 
all_input = all_input[pd.notnull(all_input['기타포괄이익누계액'])]

all_input.isnull().sum() # NaN 값이 특정 Row에 몰려 있음을 확인할 수 있다.
```

```python
# 전처리 3: Test Set 분리
# 나머지 Data 이용, 19년 11월의 월 이율 및 변동성을 예측한다

# Index에 따라 오름차순 정렬
all_input = all_input.sort_index()

# 19년 12월 정보의 첫 Index 구하기
for i, idx in enumerate(all_input.index): 
    if idx[:4] == '1911':
        # 19년 12월 기준으로 DataFrame 나누기
        test_data = all_input.iloc[i:]
        all_input = all_input.iloc[:i]
        break

# NaN 값 분포 체크
test_data.isnull().sum()
```

```python
# Test Set의 NaN 값은 다른 값으로 채울 수 없으므로 NaN 값을 갖는 Row를 제거한다
test_data = test_data.dropna(axis=0)
test_data
```

```python
# 나머지 NaN 값을 해당 Column의 평균으로 채운다
# 모델 성능 저하 요소
for column in all_input.columns:
  all_input[column].fillna(all_input[column].mean(), inplace=True)
all_input.isnull().sum()
```

```python
# 전처리 4: Data Normalization
# Column별 평균 0, 분산 1로 맞춰준다.
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# X, y 분리
y_data = all_input[['월간수익률(t+1)']]
x_data = all_input.drop(['월간수익률(t+1)'], axis='columns')

y_data.index = list(all_input.index)

# 평균 0, 분산 1로 Scaling, y_data의 경우 상관관계 분석을 위한 Frame을 따로 둔다
sscaler = StandardScaler()
sscaler.fit(x_data)

x_data = pd.DataFrame(sscaler.transform(x_data), columns=x_data.columns)
x_data.index = list(all_input.index) # Index 손실 방지

sscaler.fit(y_data)
y_data_norm = pd.DataFrame(sscaler.transform(y_data), columns=y_data.columns)
y_data_norm.index = list(all_input.index)

x_data
```

```python
# Test Data X, y Data 분리
y_test = test_data[['월간수익률']]
x_test = test_data.drop(['월간수익률'], axis='columns')

# Test Data Scaling
sscaler = StandardScaler()
sscaler.fit(x_test)

x_test = pd.DataFrame(sscaler.transform(x_test), columns=x_test.columns)
x_test.index = list(test_data.index)

x_test
```

```python
# 전처리 5 : Dataset 분리
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(x_data, y_data, test_size=0.2, random_state=1)
print(X_train.shape)
print(X_valid.shape)
print(y_train.shape)
print(y_valid.shape)
```

```python
# 모델 선택 1: Model Initialization
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import ElasticNet
from sklearn.svm import SVR
```

```python
# Random Forest
RFR = RandomForestRegressor(max_depth=7, n_estimators=200, criterion='mse', random_state=1, n_jobs=-1)

s_time = time.time()
RFR.fit(X_train, y_train)
RFR_train = RFR.predict(X_train)
RFR_valid = RFR.predict(X_valid)
print('Elapsed :', time.time() - s_time)
```

```python
# Decision Tree
DTR = DecisionTreeRegressor(max_depth=4)

s_time = time.time()
DTR.fit(X_train, y_train)
DTR_train = DTR.predict(X_train) 
DTR_valid = DTR.predict(X_valid)
print('Elapsed :', time.time() - s_time)
```

```python
# K Nearest Neighbors
KNR = KNeighborsRegressor(n_neighbors=30)

s_time = time.time()
KNR.fit(X_train, y_train)
KNR_train = KNR.predict(X_train)
KNR_valid = KNR.predict(X_valid)
print('Elapsed :', time.time() - s_time)
```

```python
# Elastic Net
ENET = ElasticNet(alpha=1.0, l1_ratio=0.5)

s_time = time.time()
ENET.fit(X_train, y_train)
ENET_train = ENET.predict(X_train)
ENET_valid = ENET.predict(X_valid)
print('Elapsed :', time.time() - s_time)
```

```python
# Gradient Boosting Regressor
GBR = GradientBoostingRegressor(max_depth=7, n_estimators=200)

s_time = time.time()
GBR.fit(X_train, y_train)
GBR_train = GBR.predict(X_train)
GBR_valid = GBR.predict(X_valid)
print('Elapsed :', time.time() - s_time)
```

```python
# Support Vector Machine Regressor
SVMR = SVR(C=100)

s_time = time.time()
SVMR.fit(X_train, y_train)
SVMR_train = SVMR.predict(X_train)
SVMR_valid = SVMR.predict(X_valid)
print('Elapsed :', time.time() - s_time)
```

```python
# Overfitting Check

from sklearn.metrics import mean_squared_error

ensemble_train = (RFR_train + DTR_train + KNR_train + ENET_train + GBR_train + SVMR_train) / 4
ensemble_valid = (RFR_valid + DTR_valid + KNR_valid + ENET_valid + GBR_valid + SVMR_valid) / 4

# y0
print('Train RMSE for y0')
print('RFR RMSE :', np.sqrt(mean_squared_error(y_train[y_train.columns[0]].tolist(), RFR_train[:,0])))
print('DTR RMSE :', np.sqrt(mean_squared_error(y_train[y_train.columns[0]].tolist(), DTR_train[:,0])))
print('KNR RMSE :', np.sqrt(mean_squared_error(y_train[y_train.columns[0]].tolist(), KNR_train[:,0])))
print('ENET RMSE :', np.sqrt(mean_squared_error(y_train[y_train.columns[0]].tolist(), ENET_train[:,0])))
print('GBR RMSE :', np.sqrt(mean_squared_error(y_train[y_train.columns[0]].tolist(), GBR_train[:,0])))
print('SVMR RMSE :', np.sqrt(mean_squared_error(y_train[y_train.columns[0]].tolist(), SVMR_train[:,0])))
print('Ensembled Model RMSE :', np.sqrt(mean_squared_error(y_train[y_train.columns[0]].tolist(), ensemble_train[:,0])))

print('\nSTD of y_train :', np.std(y_train[y_train.columns[0]].tolist()))

print('\nValid RMSE for y0')
print('RFR RMSE :', np.sqrt(mean_squared_error(y_valid[y_valid.columns[0]].tolist(), RFR_valid[:,0])))
print('DTR RMSE :', np.sqrt(mean_squared_error(y_valid[y_valid.columns[0]].tolist(), DTR_valid[:,0])))
print('KNR RMSE :', np.sqrt(mean_squared_error(y_valid[y_valid.columns[0]].tolist(), KNR_valid[:,0])))
print('ENET RMSE :', np.sqrt(mean_squared_error(y_valid[y_valid.columns[0]].tolist(), ENET_valid[:,0])))
print('GBR RMSE :', np.sqrt(mean_squared_error(y_train[y_train.columns[0]].tolist(), GBR_valid[:,0])))
print('SVMR RMSE :', np.sqrt(mean_squared_error(y_train[y_train.columns[0]].tolist(), SVMR_valid[:,0])))
print('Ensembled Model RMSE :', np.sqrt(mean_squared_error(y_valid[y_valid.columns[0]].tolist(), ensemble_valid[:,0])))

print('\nSTD of y_valid :', np.std(y_valid[y_valid.columns[0]].tolist()))
```
